### Clone GitHub Repository and Add to System Path

```python
# --- Configuration for your GitHub Repository ---
GITHUB_REPO_URL = "https://github.com/JackMclaughlin424/csci739-project-mamba.git" # Replace with your GitHub repository URL
REPO_NAME = "csci739-project-mamba"
COMMIT_HASH = None # Set this to a specific commit hash (e.g., "your_commit_hash") to check out that commit. If None, the latest commit on the specified branch will be used.
BRANCH_NAME = "experiments" # Set this to the name of the branch you want to clone (e.g., "dev", "main"). If None, the default branch will be cloned.
SUBDIRECTORY = "" # If your project code is in a subdirectory (e.g., 'Colab resources/style-prompt-generator'), specify it here. Otherwise, set to None or an empty string.
```

**Clear GitHub Repository**


```python
import shutil
import os

# Ensure REPO_NAME is defined (it should be from previous cells)
if 'REPO_NAME' in globals() and REPO_NAME:
    repo_path = f'/content/{REPO_NAME}'
    if os.path.exists(repo_path):
        print(f'Removing {repo_path}...')
        shutil.rmtree(repo_path)
        print(f'{repo_path} removed successfully.')
    else:
        print(f'Repository directory {repo_path} does not exist.')
else:
    print('REPO_NAME variable not found or is empty. Please ensure the repository name is set.')
```

**Clone repo**

```python
import os
import sys


# Construct the git clone command
clone_command = f"git clone "
if BRANCH_NAME:
    clone_command += f"--branch {BRANCH_NAME} "
clone_command += f"{GITHUB_REPO_URL} /content/{REPO_NAME}"

# Clone the repository
get_ipython().system(clone_command)

# Navigate into the cloned directory
%cd /content/{REPO_NAME}

# Checkout a specific commit if provided
if COMMIT_HASH:
    print(f"Checking out commit: {COMMIT_HASH}")
    get_ipython().system(f"git checkout {COMMIT_HASH}")
else:
    if not BRANCH_NAME:
        print("Using the latest commit on the default branch.")
    else:
        print(f"Using the latest commit on branch: {BRANCH_NAME}")

# Navigate back to the content directory (or stay in the repo root if that's where the code is)
%cd /content

# Add the repository (or its subdirectory) to sys.path
repo_path = f"/content/{REPO_NAME}"
if SUBDIRECTORY:
    repo_path = os.path.join(repo_path, SUBDIRECTORY)

if repo_path not in sys.path:
    sys.path.insert(0, repo_path)
    print(f"Added {repo_path} to sys.path")
else:
    print(f"{repo_path} is already in sys.path")

print("Current sys.path:")
for p in sys.path:
    print(p)
```
